3. Explain the use of grid-auto-row and grid-auto-column using code examples.
- grid-auto-rows: This property controls the height of new rows that the browser adds automatically when your grid items overflow.

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Grid auto rows</title>

    <style>
      .grid-container {
        display: grid;
        grid-template-columns: 100px 100px;
        grid-auto-rows: 60px;
        gap: 8px;
      }

      .grid-container div{
        background-color: bisque;
      }
    </style>
  </head>
  <body>
    <div class="grid-container">
      <div>1</div>
      <div>2</div>
      <div>3</div>
    </div>
  </body>
</html>

```

- grid-auto-columns: This property controls the width of new columns that the browser adds automatically when needed (especially in a horizontal layout).

```
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Grid auto rows</title>

    <style>
      .grid-container {
        display: grid;
        grid-auto-flow: column;
        grid-template-rows: 100px;
        grid-auto-columns: 80px;
        gap: 8px;
      }

      .grid-container div {
        background-color: bisque;
      }
    </style>
  </head>
  <body>
    <div class="grid-container">
      <div>1</div>
      <div>2</div>
      <div>3</div>
    </div>
  </body>
</html>
```
